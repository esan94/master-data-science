````python
@np.vectorize
def Id(matrix):
    '''
    
    Transform a unitary matrix to a clean unitary matrix. This function is vectorizer.
    
    :param matrix disrty unitary matrix
    return: Clean unitary matrix
    
    '''
    return 1.0 if math.isclose(matrix, 1., abs_tol=0.01) else 0.0
````

#### Exercise

In a chicken and rabbit farm, there are 35 heads and 94 legs. How many chickens and how many rabbits do we have?



Remember:

$$A \cdot X = B$$

$$A^{-1} \cdot A \cdot X = I \cdot X = A^{-1} \cdot B$$

$$X = A^{-1} \cdot B$$

````python
A = np.array([[2, 4], [1, 1]])
B = np.array([[94], [35]])
AI = np.mat(A).I
an = AI.dot(np.mat(B))
print('There are %i chickens\n' %np.array(an)[0][0])
print('There are %i rabbits\n' %np.array(an)[1][0])
````

### Filtering out missing data

any() and all() are functions of boolean Series. They reduce the Series to a single boolean value by applying repeatedly the operators "or" and "and", respectively.

The thresh argument specifies the minimum number of non-null values required to keep a column (or row, with axis=1)

````python
df3.notnull().any()
````

````python
df3.notnull().all(axis=1)
````

### Filling in missing data

````python
values = (np.random.randn(18) * 100 + 100).reshape(6, 3)
````

````python
values
````

````python
df4 = pd.DataFrame(values, index=list('abcdef'), columns=list('xyz'))
df4.iloc[1:4, 2] = np.nan
df4.iloc[3:5, 1] = np.nan
df4
````

````python
df4.dropna(thresh=2) # by row
````
