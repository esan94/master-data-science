# Saving to Excel

```python
df = pd.read_csv('914310910_T_T100_SEGMENT_ALL_CARRIER_2014_All.csv')
df.head()
```

```python
pd.options.display.max_columns = None
```

```python
df.sample(5)
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df_exercise1 = df[['PASSENGERS', 'CARRIER', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME']]
```

```python
df_exercise1.sort_values(by='PASSENGERS', ascending=False).head(10)
```

```python
df_exercise2 = df[['PASSENGERS', 'CARRIER', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME', 'DISTANCE']]
```

```python
df_exercise2['RPM'] = df_exercise2['PASSENGERS'] * df_exercise2['DISTANCE']
top10 = df_exercise2.sort_values(by='RPM', ascending=False).head(10)
```

```python
df_exercise3 = df[['PASSENGERS', 'CARRIER', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME', 'DISTANCE', 'SEATS']]
df_exercise3['SEATS'] = df_exercise3['SEATS'].replace(0.0, -1) #np.inf
df_exercise3['RPM/ASMs'] = df_exercise3['PASSENGERS'] / df_exercise3['SEATS']
df_exercise3.sort_values(by='RPM/ASMs', ascending=False).head(10)
```

```python
top10.to_excel('top10.xlsx')
```

```python
top10.to_csv('top10.csv')
```
